### Data Prepration

```python
def process_images():
    dataset={'data':None, 'target':None}

    images=[]
    targets=[]

    for class_ in range(3):
        print(class_)
        print()
        
        
        for imgname in os.listdir("./Data/"+str(class_)):
            path="./Data/"+str(class_)+"/"+imgname
            print(path)
            if imgname!=".DS_Store":
                img=cv2.imread(path)[:,:,1]
                img=cv2.resize(img,(100,100))
                images.append(img)
                targets.append(int(class_))
            else:
                pass

    dataset['data']=np.array(images)
    dataset['target']=np.array(targets)
    
    return dataset
```

```python
dataset=process_images()
```
